Matplotlib is a Python library for creating static, animated, and interactive visualizations in Python.
It is a popular and widely-used data visualization library that provides a variety of plotting functions and tools for creating high-quality charts, graphs, and other visual representations of data.

USE OF MATPLOTLIB :
1.Exploring data: Matplotlib can be used to quickly visualize data to gain insights into its underlying structure, identify trends and patterns, and spot outliers.
2.Communicating results: Matplotlib can be used to create high-quality charts and graphs for inclusion in reports, presentations, and other communication materials.
3.Developing algorithms: Matplotlib can be used to create visualizations of algorithm outputs during the development process, allowing developers to debug and refine their algorithms more effectively.

1.Line plots: Plots that show the relationship between two variables as a line connecting the data points.
2.Scatter plots: Plots that show the relationship between two variables as a collection of individual data points.
3.Bar plots: Plots that show the distribution of a categorical variable using bars of different heights or lengths.
4.Histograms: Plots that show the distribution of a numerical variable by dividing the data into a set of bins and counting the number of data points in each bin.
5.Box plots: Plots that show the distribution of a numerical variable using a box and whiskers plot, where the box represents the interquartile range of the data, the whiskers represent the minimum and maximum values within a certain range, and any points outside the whiskers are considered outliers.

SCATTER PLOT: In Scatter plot the graph is ploted in the term of dot .

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(3)
```

```python
x = 3+np.random.normal(0,2,50)
y = 3+np.random.normal(0,2,len(x))
plt.scatter(x,y)
plt.title('Scatter plot of x and y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

#QNO.3 ANS

The subplot() function in Matplotlib is used to create multiple plots (subplots) within a single figure.

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))


axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

fig.suptitle('Four Line Plots')

plt.show()
```

#QNO.4 ANS

A bar plot is a type of graph that represents data using rectangular bars. It is used to compare different categories of data and to display the differences between them.

```python
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company, profit)
plt.title("Company Profit")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()

plt.barh(company, profit)
plt.title("Company Profit")
plt.xlabel("Profit")
plt.ylabel("Company")
plt.show()
```

#QNO.5 ANS

A box plot is a type of graph used to display the distribution of a dataset. It is also known as a box and whisker plot. 

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]
labels = ['Box 1', 'Box 2']
plt.boxplot(data, labels=labels)
plt.title("Box Plot")
plt.xlabel("Box")
plt.ylabel("Value")
plt.show()
```
